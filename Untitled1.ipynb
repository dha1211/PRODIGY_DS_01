# top 10 countries for year 2022

```python
#filter data for total population
total_pop_data=df[df['Series Code']=='SP.POP.TOTL']

#sort data based on the population for 2023
sorted_population=total_pop_data.sort_values(by='2022',ascending=False)

#top 10 countries with highest population for 2023
top_10_countries=sorted_population.head(10)
print(f"Top 10 countries of total population \n{top_10_countries[['Country Name']]}")
```

# bottom 10 countries for year 2022

```python
#filter data for total population
total_pop_data=df[df['Series Code']=='SP.POP.TOTL']

#sort data based on the population for 2022
sorted_population=total_pop_data.sort_values(by="2022",ascending=False)

#bottom 10 countries with lowest population for 2022
bottom_10_countries=sorted_population.tail(10)
print(f"Bottom 10 countries of total population \n {bottom_10_countries[['Country Name']]}")
```

# top 10 countries for year 2020

```python
#filter data for total population
total_pop_data=df[df['Series Code']=='SP.POP.TOTL']

#sort data based on the population for 2023
sorted_population=total_pop_data.sort_values(by="2020",ascending=False)

#top 10 countries with highest population for 2020
top_10_countries=sorted_population.head(10)
print(f"Top 10 countries of total population\n {top_10_countries[['Country Name']]}")
```

# bottom 10 countries for year 2020

```python
#filter data for total population
total_pop_data=df[df['Series Code']=='SP.POP.TOTL']

#sort data based on the population for 2020
sorted_population=total_pop_data.sort_values(by="2020",ascending=False)

#bottom 10 countries with lowest population for 2020
bottom_10_countries=sorted_population.tail(10)
print(f"Bottom 10 countries of total population\n {bottom_10_countries[['Country Name']]}")
```

# bar plot for top 10 countries for year 2022 and 2020

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code", data=top_10_countries,palette="coolwarm")
plt.title("Top 10 countries of total population (2022)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2020",y="Country Code", data=top_10_countries,palette="coolwarm")
plt.title("Top 10 countries of total population (2020)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# bar plot for bottom 10 countries for year 2022 and 2020

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code", data=bottom_10_countries,palette="coolwarm")
plt.title("Bottom 10 countries of total population (2022)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2020",y="Country Code", data=bottom_10_countries,palette="coolwarm")
plt.title("Bottom 10 countries of total population (2020)",fontsize=10)
plt.xlabel("Total population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# countries with highest male population

```python
male_population_data=df[df['Series Code']=='SP.POP.TOTL.MA.IN']
sorted_male_population=male_population_data.sort_values(by="2022",ascending=False)
male_top_10_countries=sorted_male_population.head(10)
print(f"Top 10 countries of male population\n {male_top_10_countries[['Country Name']]}")
```

# countries with highest female population

```python
female_population_data=df[df['Series Code']=='SP.POP.TOTL.FE.IN']
sorted_female_population=female_population_data.sort_values(by="2022",ascending=False)
female_top_10_countries=sorted_female_population.head(10)
print(f"Top 10 countries of female population\n {female_top_10_countries[['Country Name']]}")
```

# top 10 countries with highest male and female population

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022",y="Country Code", data=male_top_10_countries,palette="viridis")
plt.title("Top 10 countries with highest male population(2022)",fontsize=10)
plt.xlabel("Male population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2020",y="Country Code", data=female_top_10_countries,palette="viridis")
plt.title("Top 10 countries with highest female population(2022)",fontsize=10)
plt.xlabel("Female population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

# stacked bar plot

# top 10 countries with highest male and female population

```python
merge_data=pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
merge_data["Total Population"]=merge_data["2022_male"]+merge_data["2022_female"]
```

```python
merge_data.head()
```

```python
sorted_data=merge_data.sort_values(by="Total Population",ascending=False)
```

```python
top_10_countries=sorted_data.head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data=top_10_countries,color="orange",label="female population")
sns.barplot(x="Country Code",y="2022_female",data=top_10_countries,bottom=top_10_countries["2022_female"],color="green",label="male population")
plt.xlabel("Country")
plt.ylabel("population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```

# bottom 10 countries 

```python
bottom_10_countries=sorted_data.tail(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Country Code",y="2022_female",data=bottom_10_countries,color="orange",label="female population")
sns.barplot(x="Country Code",y="2022_female",data=bottom_10_countries,bottom=bottom_10_countries["2022_female"],color="green",label="male population")
plt.xlabel("Country")
plt.ylabel("population")
plt.legend()
plt.xticks(rotation=45,ha="right")
plt.show()
```
